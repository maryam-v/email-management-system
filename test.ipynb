```python
def send_simple_email(sender, recipient, subject, body):
    # create the email message
    msg = MIMEText(body)
    msg["Subject"] = subject
    msg["From"] = sender
    msg["To"] = recipient

    # Set up the SMTP server
    smtp_server = "smtp.gmail.com"
    port = 587
    sender_email = "mvgc2025@gmail.com"
    password = os.getenv("EMAILPASS")

    try:
        with smtplib.SMTP(smtp_server, port) as server:
            server.starttls()
            server.login(sender_email, password)
            server.send_message(msg)
            print("Email sent successfully!")
    except Exception as e:
        print(f"An error occured: {e}")
```

```python
def send_html_email(sender, recipient, subject, html_content, text_content):
    msg = MIMEMultipart("alternative")
    msg["Subject"] = subject
    msg["From"] = sender
    msg["To"] = recipient

    part1 = MIMEText(text_content, "plain")
    part2 = MIMEText(html_content, "html")

    msg.attach(part1)
    msg.attach(part2)

    password = os.getenv("EMAILPASS")

    with smtplib.SMTP("smtp.gmail.com", 587) as s:
        s.starttls()
        s.login(sender, password)
        s.send_message(msg)
        print("Email sent successfully!")


html_content = """
<html>
  <body>
    <h1>Hello, World!</h1>
    <p>This is an <b>HTML</b> email sent from <i>Python</i>.</p>
    <p>Here's a list:
      <ul>
        <li>Item 1</li>
        <li>Item 2</li>
        <li>Item 3</li>
      </ul>
    </p>
    <img src="https://content.pexels.com/images/canva/ai-generated-ad/off-theme/crepuscular_rays_bursting-full.jpg" alt="An example image"
    style="width:100%; max-width:600px;">
  </body>
</html>
"""

text_content = """
Hello, World!
This is a plain text version of the HTML email sent from Python.
Here's a list:
- Item 1
- Item 2
- Item 3
"""

sender = "mvgc2025@gmail.com"
recipient = "mvgc2025@gmail.com"
subject = "HTML Email Test"

send_html_email(sender, recipient, subject, html_content, text_content)
```

```python
def add_attachemnt(msg, file_path):
    with open(file_path, "rb") as file:
        part = MIMEBase("application", "octat-stream")
        part.set_payload(file.read())
    encoders.encode_base64(part)
    part.add_header("Content-Disposition", f'attachement; filename="{os.path.basename(file_path)}"')
    msg.attach(part)
```

```python
def send_email_with_attachment(sender, recipient, subject, body, file_paths):
    msg = MIMEMultipart()
    msg["From"] = sender
    msg["To"] = recipient
    msg["Subject"] = subject

    msg.attach(MIMEText(body, "plain"))

    for file_path in file_paths:
        add_attachemnt(msg, file_path)

    password = os.getenv("EMAILPASS")
    with smtplib.SMTP("smtp.gmail.com", 587) as session:
        session.starttls()
        session.login(sender, password)
        text = msg.as_string()
        session.sendmail(sender, recipient, text)
        print("Email sent successfully")


file_paths = [
    "Attachments/Maryam_Valipour_Resume_2026.pdf",
    "Attachments/Maryam Valipour_Cover letter 2026.pdf",
]
send_email_with_attachment(
    "mvgc2025@gmail.com",
    "mvgc2025@gmail.com",
    "Email with Attachment",
    "Please find the attached file.",
    file_paths,
)
```
